```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x range (from -2pi to 2pi)
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# Define the function y = sin(x) * cos(x)
y = np.sin(x) * np.cos(x)

# Create the plot
plt.plot(x, y, label=r'y = sin(x) * cos(x)', color='blue', linestyle='-', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the function y = sin(x) * cos(x)')

# Add a legend
plt.legend()
# Show grid for better readability
plt.grid(True)
# Display the plot
plt.show()



```

```python
import matplotlib.pyplot as plt # this is the default declaration
# we are importing from the matplotlib package, pyplot is a class and
# we are aliasing it as plt

import numpy as np # We need numpy as well
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05) # generate an array from 0 to math.pi*2 with
                                  # an interval of 0.05
y = np.sin(x)  # built in trignometry function in numpy
plt.plot(x,y)  # an x, y plot
plt.xlabel("angle θ")
plt.ylabel("sine")
plt.title('sine wave')
plt.grid(True)
plt.show() # the plot is shown only when you give the show command

import numpy as np # We need numpy as well
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05) # generate an array from 0 to math.pi*2 with
                                  # an interval of 0.05
y = np.cos(x)  # built in trignometry function in numpy
plt.plot(x,y, label=r'y = cos(x)', color='blue', linestyle='-', linewidth=2)  # an x, y plot
plt.xlabel("angle θ")
plt.ylabel("Cosine")
plt.title('Cos wave')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt  # Importing the plotting library
import numpy as np  # Importing numpy for numerical calculations

# Define the range of x values
x = np.linspace(-10, 10, 400)  # Create an array from -10 to 10 with 400 points

# Define the equation y = 3x^2 + 5x + 10
y = 3 * x**2 + 5 * x + 10

# Plot the equation
plt.plot(x, y, label=r'y = 3x^2 + 5x + 10', color='pink', linestyle='-', linewidth=2)  # Line plot with blue color and solid line style

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of y = 3x^2 + 5x + 10')

plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,axes =  plt.subplots(1,2) # subplots(rows, columns), this function returns
                              # two values, fig and axes, we are interested
                              # in axes

import numpy as np
x = np.arange(1,5,0.2)        # an array with values 1, to  with 0.2 steps
axes[0].plot(x, x**2 + 5*x + 12)        # first plot function at (0,0)
axes[0].set_title('x^2 + 5x + 12')
axes[1].plot( x,np.log10(x))  # plot at (0,1)
axes[1].set_title('log')

plt.tight_layout()   # spaces out the plots

plt.show()
```
