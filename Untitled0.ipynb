```python
# Let's try to print the Obama image using the three colour channels seperately
# Lets try another image
import matplotlib.pyplot as plt
import matplotlib.image as img # take image class of matplotlib as img
import numpy as np
image = img.imread('ai-drawing.jpg') # please load this image to your folder
fig,axes =  plt.subplots(2,3,figsize = (7,7))

# We can take the different colour channels

# cmap are colour maps see https://matplotlib.org/tutorials/colors/colormaps.html

axes[0][0].imshow(image[:,:,0],cmap='Reds') # Red Channel
axes[0][1].imshow(image[:,:,1],cmap='Greens') # Green Channel
axes[0][2].imshow(image[:,:,2],cmap='Blues') # Blue Channel
axes[1][0].imshow(image[:,:,0],cmap='gray') # Red Channel in gray colour
axes[1][1].imshow(image[:,:,1],cmap='gray') # Green Channel in gray colour
axes[1][2].imshow(image[:,:,2],cmap='gray') # Blue Channel in gray colour

plt.tight_layout()   # spaces out the plots

plt.show() # display all the plots
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img  # Importing the image module
import numpy as np

# Read the first image (obama.jpg)
image = img.imread('obama.jpg')  # Ensure 'obama.jpg' is in the same folder

# Read the second image (modi.jpg)
image2 = img.imread('modi.jpg')  # Ensure 'modi.jpg' is in the same folder

# Create a subplot with 1 row and 2 columns
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # (1 row, 2 columns)

# Display the first image on the first subplot (ax[0])
ax[0].imshow(image)
ax[0].axis('off')  # Turn off axis for better image display
ax[0].set_title('Image 1: obama.jpg')
print(f"Image 1 (Obama): {image.shape}")

# Display the second image on the second subplot (ax[1])
ax[1].imshow(image2)
ax[1].axis('off')  # Turn off axis for better image display
ax[1].set_title('Image 2: modi.jpg')
print(f"Image 2 (Modi): {image2.shape}")

# Show both images
plt.tight_layout()  # Adjust layout to avoid overlapping titles and images
plt.show()
```

```python
import matplotlib.pyplot as plt  # Importing the plotting library
import numpy as np  # Importing numpy for numerical calculations

# Define the range of x values
x = np.linspace(-10, 10, 400)  # Create an array from -10 to 10 with 400 points

# Define the equation y = 3x^2 + 5x + 10
y = 3 * x**2 + 5 * x + 10

# Plot the equation
plt.plot(x, y, color='blue', linestyle='-', linewidth=2)  # Line plot with blue color and solid line style

# Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of y = 3x^2 + 5x + 10')

# Display the plot
plt.grid(True)  # Adds grid for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,axes =  plt.subplots(2,3) # subplots(rows, columns), this function returns
                              # two values, fig and axes, we are interested
                              # in axes

import numpy as np
x = np.arange(1,5,0.2)        # an array with values 1, to 5 with 0.2 steps
axes[0][0].plot(x,x*x)        # first plot function at (0,0)
axes[0][0].set_title('square')
axes[0][1].plot( x,np.log10(x))  # plot at (0,1)
axes[0][1].set_title('log')
axes[0][2].plot(x,np.exp(x))  # plot at (0,2)
axes[0][2].set_title('exp')
axes[1][0].plot(x,np.sqrt(x)) # plot at (1,0)
axes[1][0].set_title('square root')
axes[1][1].plot(x, 2*x + 5)    # plot at (1,1)
axes[1][1].set_title("Linear")
axes[1][2].plot(x, 3*x**2 - 12*x + 12) # plot at (1,1)
axes[1][2].set_title("Polynomial")
plt.tight_layout()   # spaces out the plots

plt.show() # display all the plots
```
